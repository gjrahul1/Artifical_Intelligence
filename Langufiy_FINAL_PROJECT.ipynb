```python
#Language is Tamil
import conllu
with open("/content/ud-treebanks-v2.9/UD_Tamil-TTB/ta_ttb-ud-test.conllu", mode="r", encoding="utf-8") as data:
  annotations=data.read()
```

```python
ud_pos=[]
ud_features=[]
sentences= conllu.parse(annotations)
for i in range (len(sentences)):
  for j in range (len(sentences[i])):
    ud_features.append(sentences[i][j]['form']) #Creating feature list
    ud_pos.append(sentences[i][j]['upos']) #Creating POS list
```

```python
#Further, using feature list for pos tagging in nltk
print(ud_features)
```

```python
#POS Tagging
postag=nltk.pos_tag(ud_features, tagset='universal')
```

```python
#OUTPUT OF POS TAGGING
print(postag)
```

```python
#zipping features and pos tags of UD
ud_tag=list(zip(ud_features,ud_pos))
print(ud_tag)
```

```python
#Counts the range of the similar/equal POS between NLTK POS TAGGING and UD Tag of TAMIL words.
#This line of code compare the POS of the Tamil word to the 'Original Data', if the length of POS of given word is equal to the UD_Word then Count it as +1

count=0
for  i in range(len(postag)):
  if(postag[i]==ud_tag[i]):
    count+=1
```

```python
#This line of code, aims to print the accuracy of the length of the POS of the given tamil words to the actual given word for that POS, over the length of POS of Tamil Words.
#Accuracy is 23%
print("The accuracy is")
print(count/len(postag))
```
