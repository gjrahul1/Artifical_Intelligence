```python
#Creating dataframe
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df
```

* Let's complete the table

```python
df['target'] = digits.target
df.head(6)
```

```python
#Creating objects for each target

df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
df3 = df[df.target==3]
df4 = df[df.target==4]
df5 = df[df.target==5]
df6 = df[df.target==6]
df7 = df[df.target==7]
df8 = df[df.target==8]
df9 = df[df.target==9]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target'],axis=1)
Y = df.target
```

```python
X.head(5)
```

```python
Y.head(5)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(kernel="linear",C=6,gamma=10)
```

```python
model.fit(X_train,y_train)
```

```python
#Accuracy of the model
model.score(X_test,y_test)
```

Normal Accuracy: 0.9944444444444445

Accuracy with kernal('rbf'): 0.9944444444444445

Accuracy with kernal('rbf') and regularization (C = 2) : 0.9888888888888889



### Summary

* Fine Tuning model with regularization or gamma is leading to lesser accuracy of the model

* kernal ('rbf') with regularization (1/2) and gamma = 4 gives the lowest accuracy of 0.08
